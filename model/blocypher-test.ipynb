```python
li_1=[]
for i in li:
    if i.split("_")[2]=="angry.wav":
        li_1.append((i,0))
    if i.split("_")[2]=="neutral.wav":
        li_1.append((i,1))
    if i.split("_")[2]=="disgust.wav":
        li_1.append((i,2))
    if i.split("_")[2]=="sad.wav":
        li_1.append((i,3))
    if i.split("_")[2]=="fear.wav":
        li_1.append((i,4))
    if i.split("_")[2]=="happy.wav":
        li_1.append((i,5))
    if i.split("_")[2]=="ps.wav":
        li_1.append((i,6))
    
```

```python
sum=[0,0,0,0,0,0]

for i in li_1:
    if i[1]==0:
        sum[0]=sum[0]+1
    if i[1]==1:
        sum[1]=sum[1]+1
    if i[1]==2:
        sum[2]=sum[2]+1
    if i[1]==3:
        sum[3]=sum[3]+1
    if i[1]==4:
        sum[4]=sum[4]+1
    if i[1]==5:
        sum[5]=sum[5]+1
    

print(sum)
        
```

```python
def normalize(x):
    arr=[]
    for i in range(x.shape[1]-1):
        x[str(i)]=(x[str(i)]-np.mean(x[str(i)]))/np.std(x[str(i)])
        arr.append((np.mean(x[str(i)]),np.std(x[str(i)])))
    return arr,x
    
```
